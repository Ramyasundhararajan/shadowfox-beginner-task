1.Matplotlib:
    
    Matplotlib is a versatile library for creating static,animated and interactive visualizations in Python.
    It was developed by John D. Hunter in 2003.
    It offers a wide range of plots including line plots,scatter plots,bar plots,histograms,pie charsts,3D plots and more.
    It provides extensive customization options to create visually appealing and informative plots.
   
 Unique Features:
        
    1.Versatility
    2.Customization
    3.Integration with Numpy 
    4.Publication Quality
    5.Extensible 
    6.Cross Platform 
    7.Interactive Plots
 
    Matplotlib gives users granular control over every aspect of a plot,from colors and labels to axes and annotations.
    It is well-integrated with Numpy and Pandas, making it suitable for data analysis and scientific computing.
    
 Basic Components of Figure
        
    1.Figure : In Matplotlib,figure is the top-level container that holds all the elements of a plot.
    It represents the entire window or page where the plot is drawn.
        
    2.Axes : In Matplotlib,axes are the rectangular areas with the figure where data is plotted.Each figure can contain one or more axes,arranged in rows and columns.It provide the coordinate system and where most of the plotting occurs.
      
    3.Axis : In Matplotlib,axis represents x-axis and y-axis of the plot. It define the data limits,tick locations,tick labels, and axis labels.
        
    4.Marker : In Matplotlib,markers are symbols used to denote individual data points on a plots. Markers are often used in scatter plots to visually distinguish between different data points.
       
    5.Adding lines to Figures: Lines connect data points on a plot.They represent the relationship or trend between data points and can be styled with different colors,widths,and styles to convey additional information.
        
    6.Title: Title is a text element that provide the description about the plot.It appears on top of the figures and provides the information or context of the visualized data.
        
    7.Axis Labels: Labels are text elements that provides descriptions for X and Y axis.
        
    8.Ticks : Tick marks are small marks along the axis that indicate specific data points or intervals.
        
    9.Tick Labels: Tick Labels are text elements that provide labels for the tick marks.
        
    10.Legend : Legends provide a key to the symbols or colors used in the plot to represent different data series or categories. 
        
    11.Grid Lines :  Grid lines are horizontal and vertical lines that extend across the plot, corresponding to specific data intervals or divisions.
        
    12.Spines : Spines are the lines that form the borders of the plot area. They separate the plot from the surrounding whitespace and can be customized to change the appearance of the plot borders.
    
 Typical Use Case :
    
    Matplotlib is commonly used for exploratory data analysis, scientific plotting,publication-quality graphs and creating custom visualizations in research.

Line Chart:

    A line plot represents data points connected with straight lines.
    Use Case: Showing trends or changes over time.

```python
#Code :
import matplotlib.pyplot as plt
plt.subplot(3,3,1)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y,'m--')
plt.show()

#Adding markers to line plots 

import matplotlib.pyplot as plt
plt.subplot(3,3,2)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, color = "blue", linestyle = 'solid',marker = 'o',markerfacecolor = 'red')
plt.show()


#Adding grid lines in lines plots 

import matplotlib.pyplot as plt
plt.subplot(3,3,3)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y,'m--')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()


#Plotting horizontal and vertical line
import matplotlib.pyplot as plt
plt.subplot(3,3,4)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.plot(y,x)
plt.show()

#Plotting multiple lines 
import matplotlib.pyplot as plt
import numpy as np
plt.subplot(3,3,5)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y,label = "line 1")
plt.plot(y,x,label = "line 2")
plt.plot(x,np.sin(x),label = "curve 1")
plt.plot(y,np.cos(y),label = "curve 2")
plt.legend()
plt.show()


#plotting with line width
import matplotlib.pyplot as plt
plt.subplot(3,3,6)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y,linewidth = 5)
plt.show()
```

Pie Chart:
    
    Pie Charts display data as slices of a circular chart.
    Use Case: Showing proportions or percentages of a whole.

```python
#Code

import matplotlib.pyplot as plt
plt.subplot(2,2,1)
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 25, 15]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()


#setting border for wedges 

import matplotlib.pyplot as plt 
plt.subplot(2,2,2)
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 25, 15]
plt.pie(sizes, labels = labels, autopct = '%1.1f%%', 
                startangle = 90, 
                wedgeprops = {"edgecolor" : "black", 
                              'linewidth': 2, 
                               'antialiased': True}) 
plt.axis('equal') 
plt.show() 


#shadow of the piechart

import matplotlib.pyplot as plt 
plt.subplot(2,2,3)
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 25, 15]
plt.pie(sizes, labels = labels, autopct = '%1.1f%%', shadow = True,
                startangle = 90, 
                wedgeprops = {"edgecolor" : "black", 
                              'linewidth': 3, 
                               'antialiased': True}) 
plt.axis('equal') 
plt.show() 

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.random.rand(100)
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
data = np.random.randn(1000)
data_matrix = np.random.rand(10, 10)
x_area = np.linspace(0, 10, 100)
y1 = np.sin(x_area)
y2 = np.cos(x_area)

# Line Plot
plt.figure(figsize=(10, 6))
plt.subplot(3, 3, 1)
plt.plot(x, y, label='Sine Wave')
plt.title('Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

# Scatter Plot
plt.subplot(3, 3, 2)
plt.scatter(x, z, color='r')
plt.title('Scatter Plot')
plt.xlabel('X axis')
plt.ylabel('Z axis')

# Bar Plot
plt.subplot(3, 3, 3)
plt.bar(categories, values, color=['blue', 'green', 'red', 'purple'])
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Histogram
plt.subplot(3, 3, 4)
plt.hist(data, bins=30, color='c', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')


# Pie Chart
plt.subplot(3, 3, 5)
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple'])
plt.title('Pie Chart')

# Area Plot
plt.subplot(3, 3, 6)
plt.fill_between(x_area, y1, alpha=0.5, label='Sine', color='blue')
plt.fill_between(x_area, y2, alpha=0.5, label='Cosine', color='red')
plt.title('Area Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

# Contour Plot
X, Y = np.meshgrid(x, x)
Z = np.sin(X) * np.cos(Y)
plt.subplot(3, 3, 7)
plt.contour(X, Y, Z, cmap='plasma')
plt.title('Contour Plot')

plt.tight_layout()
plt.show()

#3D plot 
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Plot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

2.Seaborn :
    
    Seaborn is python library for creating statistical graphics.
    It builds on top of matplotlib and integrates closely with pandas data structures.
    
Unique Features :
    
    Statistical Visualization :  Seaborn specializes in making complex statistical graphs. It provides default styles and color palettes to enhance the visual appeal in plots.
        
    Data Exploration : Seaborn is used to explore and understand the data.It operates on dataframes and arrays.And internally performs semantic mapping and statistical aggregation to produce informative plots.
        
    Declarative API : It allows you to focus on the meaning of different plot elements rather than the details of how to draw.
        
    Beautiful Defaults: Seaborn comes with appealing default styles and color palettes, easily customize these styles to match the preferences.
        
    Integration with Pandas: Seaborn directly accepts Pandas DataFrames, allowing you to create plots directly from your data without extensive data manipulation.

Differnt Categories of Plots :
    
    Built-in Plot Types: Seaborn provides functions for common plot types, including scatter plots, line plots, bar plots, histograms, kernel density estimation (KDE) plots, violin plots, and more.

    Statistical Aggregation: Many Seaborn functions internally perform statistical aggregation, such as computing means, confidence intervals, or other summary statistics. This makes it easy to visualize relationships between variables.

    Pair Plots and Heatmaps: Seaborn offers convenient pair plots (scatter matrix plots) and heatmaps for exploring relationships between multiple variables.

    Categorical Data Visualization: Seaborn excels at visualizing categorical data, including box plots, violin plots, and categorical scatter plots.

    Regression Plots: Seaborn provides regression plots (like linear regression and logistic regression) with confidence intervals, which can be useful for understanding relationships between variables.
 
    Themes and Contexts: We can easily change the overall theme and context of our plots using Seaborn’s built-in options.
    
    Additional Plot Customization: Seaborn allows fine-tuning of plot aesthetics, including color palettes, grid styles, and font sizes.
        
    Matrix plots: A matrix plot is an array of scatterplots.
        
    Multi-plot grids: It is a useful approach to draw multiple instances of the same plot on different subsets of the dataset.
    
    
Use Case :
    
    Exploratory Data Analysis (EDA): Quickly visualize data distributions, correlations, and patterns.
    Comparing Groups: Compare distributions across different groups using box plots, violin plots, or bar plots.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
a = np.random.normal(size=1000)
tips = sns.load_dataset("tips")

# Create a 3x3 grid for subplots
plt.figure(figsize=(12, 10))

# Histplot
plt.subplot(3, 3, 1)
sns.histplot(a, kde=True, color="y")
plt.title("Histplot")

# Distplot
plt.subplot(3, 3, 2)
sns.distplot(a, kde=True, color="b")
plt.title("Distplot")

# Lineplot
plt.subplot(3, 3, 3)
sns.lineplot(data=tips, x="total_bill", y="tip", color="y")
plt.title("Lineplot")

# Lmplot (regression plot)
plt.subplot(3, 3, 4)
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker")
plt.title("Lmplot")

# Stripplot
plt.subplot(3, 3, 5)
sns.stripplot(data=tips, x="day", y="total_bill", jitter=True)
plt.title("Stripplot")

# Swarm plot
plt.subplot(3, 3, 6)
sns.swarmplot(data=tips, x="day", y="total_bill")
plt.title("Swarm Plot")

# Bar plot
plt.subplot(4, 4, 9)
sns.barplot(data=tips, x="day", y="total_bill", ci=None)
plt.title("Bar plot")

# Box plot
plt.subplot(4, 4, 10)
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Box Plot")

# Count plot
plt.subplot(4, 4, 11)
sns.countplot(data=tips, x="day")
plt.title("Count Plot")

# Violin plot
plt.subplot(4, 4, 12)
sns.violinplot(data=tips, x="day", y="total_bill")
plt.title("Violin Plot")

plt.tight_layout()
plt.show()
```

Comparison between Matplotlib ans seaborn :

Functionality:
        Matplotlib: It is used for creating basic graphs like bar charts, histograms, pie charts, scatter plots, and more. It            provides a wide range of customizable options and acts as a graphics package for data visualization in Python.
        Seaborn: Built on top of Matplotlib, Seaborn simplifies statistical plots with built-in themes. It offers more advanced          visualizations with less code, making it great for univariate and bivariate data exploration.
        
        
Syntax:
        Matplotlib: Uses comparatively complex and lengthy syntax. For example, creating a bar graph:                          n matplotlib.pyplot.bar(x_axis, y_axis).                         
        Seaborn: Has simpler syntax, making it easier to learn and understand. For a bar graph: seaborn.barplot(x_axis, y_axis).


Dealing with Multiple Figures:
        Matplotlib: Allows opening and using multiple figures simultaneously, which can be closed individually or all at once.
        Seaborn: Sets the time for creating each figure, but be cautious about memory issues (OOM) when dealing with large 
         datasets.


Visualization:
        Matplotlib: Well-connected with Numpy and Pandas, it’s suitable for various plot types.
        Seaborn: More comfortable with Pandas data frames, and it avoids overlapping plots using default themes.
